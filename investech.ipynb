## 第一节 导入j0和topdown指标 以及两指标的含义
* 按快捷键ctrl+f打开公式管理器面板，点击导入公式，导入.tne格式文件，j0和topdown两个指标，确定。 j0+enter，topdown+enter让这两指标显示在图上。 公式管理器面板上，查找jo和topdown指标，双击指标，进入指标公式编辑器，通达信软件底下就可以查看两个公式涵义。 
* 那么我简单说下两个指标的意思。 m10,m30,m160 这个很简单，计算的就是10日均线，30日均线和160日均线 
jo一共有5根线，HH（默认为浅蓝色虚线）最顶上计算的是空头防线的最优成本线，空头是靠借钱买入股票在高点卖出赚钱的，他计算的是往前去看，我们往前看行情，出现个最高点开始，往后每天都有一个最高点去算一个平均值连成的这条线，底下则是多头市场的最佳成本线，中间取二分之一的MM（白色）相当于楚河汉界了。然后上中，中下各取二分之一形成两根线HP和LP（蓝色）。 topdown指标表示振幅，具体计算方法可以查看公式。他是按最高价计算的，如果最高价>收盘价，则返回最高价，如果最高价<收盘价，则返回收盘价实际上他就是幅度，就是通达信条件太少了，他不能完整的把一段的幅度计算出来，那现在算出来的是一段当中的一小片一小片的幅度。但是他要考虑两种情况，一个是在高位还有一个是在低位，所以有时候他选择最高价有时候是选收盘价，如果是在低位的时候算今天的收盘价跟之前的低点又相差多少的话，那就用收盘价，因为他今天是下跌的，收盘价会低于最高价很多对不对，那用最高价就没道理了，如果他是上涨的只是回调那你用最高价就可以，所以他里边有一句话用的来说明是选最高价还是收盘价。

## 第三节 什么是空头市场和多头市场
空头市场是指在一段行情中，价格不再创新高，而反复创新低，最近价格不再突破之前最高价。空头市场中，天线是比较连续的，而地线则是断断续续的，不断被向下突破。
多头市场则相反。
比如我们举个例子，假如一个人减肥，如果他的体重不再创历史新高了，就说明减肥是比较成功了。假如体重一直创新高，那么就不是减肥了。这个应该是很好理解的！

## 第四节 投资品种的确定性
* 1.静态确定性比如分级a，他的上下是较为固定的。
* 2动态确定性 一个品种他在不同时期或者向位他具有一些较为确定性的特点。举个例子，比如大雁到了冬天就往南飞，会根据季节变化而变化，有一定的规律。我们结合j0指标去看发现，股价的范围基本都是在空头压制线和多头压制线以内的，那么这个就是说他是有一定的确定性的。

## 第六节 多头市场买入时机
** 多头市场的买入时机：**
* 一，td到0往上
* 二，结合j0指标去看
在多头市场中，如果他踩住空头防线对其有较强支撑。多头市场当中是天线最保险，天线下面那个中间线其次保险。中间的白线為多空的分界线，不保险。因為显示多头力量不够。在上面活动的时候多头强劲，在下面活动的时候空头强劲。对了最下面的那根线的红色的呢就是多头的老本营了，退到了25处可退了，纸能是往下逃跑了所以一般在那个地方会拼命的挣扎著，那如果不退到那个地方或者说很难冲过去，副地线，再也不回来。那就是多头稳住了。有一根黄色虚线，是天线当中的地线。因為它可以重重叠叠大中套小，它是可以嵌套起来的。
就是说，假如我们把天线之前的行情都不看的话。那根线是算出来的底线。黄线对应著红线。一个是红线是总的底线。黄线是天线以后这一段行情的底线。这样重重叠叠还可以一直算下去的。d线支撑的天线天线最终的底线将重重叠叠。
就是说这个天线和地线可以重重叠叠的算下去的，就看取值的范围，我从整个行情去看，蓝色和红色这是两个最外边的边框。我们可以这样，取一段行情之前的不去看，那么从这里起我们再按照j0天线的计算方法再计算可以得到一天新的天线，地线也是一样的。
踩天线和被天线盖住是不一样的。其实就相当于两军打仗，当多头打胜仗，空头防线被突破后，实际上这个力量是很强势的，相当于你败方大集体被打败了，需要换地扎营了，剩余的士兵也将很快退却。

## 第七节  卖出时机
**卖出时机通用方法**
* 1，根据不同週期的k线综合判断趋势线跌破卖出
* 2，10日均线跌破卖出涨幅达到统计学大概率卖出区间
* 3.在多头或空头市场中,明显遇到天线类阻挡,尤其是在空头市场中遇到任何天地线阻挡
* 4.已经涨5%以上,又看不清行情趋势,分次止盈卖出 
* 5.没有涨多少掉头亏损，例如亏1%，止损卖出。
* 6弯折可卖，td弯折，均线弯折

## 第八节 分级基金的买卖循环
**上海分级基金更加特殊的循环方法(T+0)**
* 当分级A出现0下回昇且符合其他买入条件时,买入分级A,若持有分级母基金或B则卖出
* 当分级母基金例如军工分级从0下回昇且符合其他买入条件时，买入母基金(5万份以上才可以拆分)，拆分為AB，随即卖出分级A
* 如果分级b离涨停很远则跟随其上涨,待超过买入点时即可考虑卖出赚差价
* 如果分级b已经涨停,这时账面上看似亏损,原因是分级b没法再涨,等次日补涨再卖深沪两市分级基金
